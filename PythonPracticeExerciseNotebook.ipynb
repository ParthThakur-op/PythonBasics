### Excercise : 22

Create program where the user can enter a number , and _if_ the number is even print `'This number is even!'` or _else_ print `'This number is odd'` and otherwise `'This is not a valid number'`

```python
number = float(input('Enter your number: '))
    

if number %2 == 0 :
    print('number is even')
elif number %2 != 0 :
    print('number is odd')
else:
    print('This is not a valid number')
```

```python
try:
    number = float(input("Enter your number: "))

    if number.is_integer() == False:
        print("This is not a valid number")
    elif int(number) % 2 == 0:
        print("This number is even!")
    else:
        print("This number is odd")

except ValueError:
    print("This is not a valid number")
```

```python
number = float(input("Enter your number: "))

if number != int(number):
    print("This is not a valid number")
elif number % 2 == 0:
    print("This number is even!")
else:
    print("This number is odd!")
```

```python
# Extract all the odd numbers from the given list

li = [2, 6, 5, 4, 7, 8, 6, 9, 11, 32, 45, 44, 67, 87, 90, 91]

odd_numbers = [num for num in li if num % 2 != 0]

print(odd_numbers)
```

```python
# Count vowels and consonants

string = "What a wonderful day!"
v = c = 0

for ch in string.lower():
    if ch in "aeiou":
        v += 1
    elif ch.isalpha():
        c += 1

print("Vowels =", v)
print("Consonants =", c)
```

```python
# Find the largest word

sentence = "What a wonderful day!"

words = sentence.split()
largest = max(words, key=len)
print(words)

print("Largest word =", largest)
```

```python
# another way 
sentence = "What a wonderful day!"

words = sentence.split()
largest = ""

for word in words:
    if len(word) > len(largest):
        largest = word

print("Largest word =", largest)
```
